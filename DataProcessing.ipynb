```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2

data_path = 'Data'
training = os.path.join(data_path, 'Training')
testing = os.path.join(data_path, 'Testing')

classes = ['Healthy', 'Infested', 'Dead']
class_t = 2


train_path = os.path.join(training, classes[class_t])
files = os.listdir(train_path)

random_idx = np.random.randint(0, len(files))
random_file = os.path.join(train_path, files[random_idx])
patch_cv = cv2.imread(random_file)
patch = imread(random_file)

fig = plt.figure()
plt.subplot(2, 2, 1)
plt.imshow(patch[1:-1, 1:-1, :])
plt.subplot(2, 2, 3)
plt.imshow(patch_cv[1:-1, 1:-1, :])

test_path = os.path.join(testing, classes[class_t])
files = os.listdir(test_path)

random_idx = np.random.randint(0, len(files))
random_file = os.path.join(test_path, files[random_idx])
patch_cv = cv2.imread(random_file)
patch = imread(random_file)

plt.subplot(2, 2, 2)
plt.imshow(patch[1:-1, 1:-1, :])
plt.subplot(2, 2, 4)
plt.imshow(patch_cv[1:-1, 1:-1, :])
print(patch[1, 1, :-1] * 255, patch_cv[1, 1, :])
```
